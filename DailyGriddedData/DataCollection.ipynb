```python
import pickle
import pandas as pd
import numpy as np
import random
```

```python
def validate(l,threshold=5, sampleSize=10, allEntry=False):
    # Function used to validate that the results are non-zero entries
    # To be used in future to filter out all zero entries

    if allEntry:
        return any(l)
    
    x=random.sample(l,sampleSize)
    cnt=0
    for i in x:
        if i!=float(0):
            cnt+=1
    
    return cnt>=threshold
```

```python
DIR="./coordInterval3/"

count=0
for i in range(0,90,3):
    lat=str(i)
    for j in range(0,180,3):
        lng=str(j)
        
        for k in range(4):
            if k==0:
                fileName=lat+"N-"+lng+"E.pkl"
            elif k==1:
                fileName=lat+"N-"+lng+"W.pkl"
            elif k==2:
                fileName=lat+"S-"+lng+"E.pkl"
            elif k==3:
                fileName=lat+"S-"+lng+"W.pkl"

            
            loadFile = open(DIR+fileName, 'rb')
            l = pickle.load(loadFile)
            if validate(l,5,10,False):
                results[fileName.strip(".pkl")]=l
#             else:
#                 count+=1
#                 print("Not Valid: ", fileName)
```

```python
# Corrected coord
DIR="./coordInterval3/"
resultsInterval3=dict()

count=0
for i in range(0,90,3):
    lat=str(i)
    for j in range(0,180,3):
        lng=str(j)
        
        for k in range(4):
            if k==0:
                fileName=lat+"N-"+lng+"E.pkl"
                keyName=str(i+0.5)+"N-"+str(j+0.5)+"E"
            elif k==1:
                fileName=lat+"N-"+lng+"W.pkl"
                keyName=str(i+0.5)+"N-"+str(j-0.5)+"E"
            elif k==2:
                fileName=lat+"S-"+lng+"E.pkl"
                keyName=str(i-0.5)+"N-"+str(j+0.5)+"E"
            elif k==3:
                fileName=lat+"S-"+lng+"W.pkl"
                keyName=str(i-0.5)+"N-"+str(j-0.5)+"E"

            
            loadFile = open(DIR+fileName, 'rb')
            l = pickle.load(loadFile)
            if validate(l,7,10,False):
                resultsInterval3[keyName]=l
```
